## Умножение матриц

Обозначения:

- $X$ — матрица признаков (нулевой столбец состоит из единиц)

- $y$ — вектор целевого признака

- $P$ — матрица, на которую умножаются признаки

- $w$ — вектор весов линейной регрессии (нулевой элемент равен сдвигу)

Предсказания:

$$
a = Xw
$$

Задача обучения:

$$
w = \arg\min_w MSE(Xw, y)
$$

Формула обучения:

$$
w = (X^T X)^{-1} X^T y
$$

**Ответ:** Качество линейной регрессии при умножении признаков на обратимую матрицу не изменится.

**Обоснование:**
Представим новую матрицу признаков X1 как произведение старой X на матрицу P(обратимую):
$$X1 = X∗P$$
Подставим новое значение X1 в формулу w1:
$$w1=((XP)^TXP)^{−1}(XP)^Ty$$
Раскроем первое произведение (XP)^T:
$$w1=(P^TX^TXP)^{-1}P^TX^Ty$$
Перегруппируем множители в скобках, чтобы это выглядело как произведение трех множителей:
$$w1=(P^T(X^TX)P)^{-1}P^TX^Ty$$
Раскроем скобки $(P^T(X^TX)P)^{-1}$ :
$$w1=P^{-1}(X^TX)^{-1}(P^T)^{-1}P^TX^Ty$$
Так как $P$ по условию обратимая, то произведение  $(P^T)^{−1}P^T$ равно $E$(единичная матрица):
$$w1=P^{-1}(X^TX)^{-1}EX^Ty=P^{-1}(X^TX)^{-1}X^Ty$$
Можно заметить, что справа получилась формула для $w$:
$$w1=P^{-1}(X^TX)^{-1}X^Ty=P^{-1}w$$
Подставим новое значение весов $𝑤1$ в формулу для предсказаний линейной регрессии. Предсказания модели расчитываются по формуле:
$$a=Xw$$
Подставим в эту формулу значения $X1$ и $w1$ для того, чтобы на их основе вычислить предсказания модели $a1$:
$$a1=X1w1=XPP^{-1}w$$
Так как $P$ по условию обратимая, то произведение $PP^{−1}$ снова равно $E$:
$$a1=XPP^{-1}w=XEw=Xw=a$$
Мы доказали, что предсказания $a1$ для матрицы признаков, умноженных на обратимую матрицу $P$ равны предсказаниям $a$.
В данном случае, параметры линейной регрессии в исходной задаче $w$ и в преобразованной $w1$ связаны следующим образом:
$$w1=P^{-1}w$$

### Разбивка датасета на наборы данных

```python
def data_split(datasets,
               test_sets=False,
               train_size=0.7,
               valid_size=0.3,
               target_name='',
               random_state=RANDOM_STATE):
    '''Функция делит датасеты на выборки:
        - обучающая
        - валидационная
        - тестовая

    Args:
        datasets: (array) Массив датасетов
        test_sets: (bool) Необходима ли тестовая выборк, по умолчанию False
        train_size: (float) Размер доли обучающей выборки, по умолчанию 0.7
        valid_size: (float) Размер доли валидационной выборки, по умолчанию 0.3. Этим параметром управляется размер тестовой выборки по остаточному принципу
        target_name: (string) Наименование столбца целевого признака
        random_state: (int or object numpy random_stat)

    Returns:
        array: Упакованные массивы с выборками.'''
    
    datasets_jointed = []

    if ((round(train_size + valid_size, 2) == 1) & (test_sets == True)) | ((train_size + valid_size) != 1.0) & (test_sets == False):
        print('Проверь размер датасетов ')
    else:
        if test_sets == False:
            for set in datasets:
                print(f'___________________________')
                target_all = set[target_name]
                features_all = set.drop(target_name, axis=1)

                features_train, features_valid, target_train, target_valid = train_test_split(
                    features_all, target_all, test_size=valid_size, train_size=train_size, random_state=random_state)

                print(f'target_train размер: {len(target_train) / len(target_all) * 100:.0f}%')
                print(f'target_valid размер: {len(target_valid) / len(target_all) * 100:.0f}%')
                display(features_train.head())
                datasets_jointed.append([features_train, features_valid, target_train, target_valid])
        else:
            for set in datasets:
                print(f'___________________________')
                target_all = set[target_name]
                features_all = set.drop(target_name, axis=1)

                test_s = round(1 - train_size - valid_size, 2)
                valid_size_prop = round(valid_size / (train_size + valid_size), 2)

                features, features_test, target, target_test = train_test_split(
                    features_all, target_all, test_size=test_s, train_size=1-test_s, random_state=RANDOM_STATE
                )
                features_train, features_valid, target_train, target_valid = train_test_split(
                    features, target, test_size=valid_size_prop, train_size=1-valid_size_prop, random_state=RANDOM_STATE
                )

                print(f'target: {target_name}')
                print(f'target_train размер: {len(target_train) / len(target_all) * 100:.0f}%')
                print(f'target_valid размер: {len(target_valid) / len(target_all) * 100:.0f}%')
                print(f'target_test размер: {len(target_test) / len(target_all) * 100:.0f}%')
                display(features_train.head())
                datasets_jointed.append([features_train, features_valid, features_test, target_train, target_valid, target_test])
    return datasets_jointed
```

### Преобразование наборов данных в матрицы
Библиотека Numpy работает с матрицами, и при помощи этой библиотеки можно проводить любые преобразования матриц. Переходим от использования Pandas к Numpy.

```python
features_train_matrix = features_train.values
features_test_matrix = features_test.values

print(features_train_matrix.shape)
print(features_test_matrix.shape)
```

### Рандомная обратимая матрица

```python
def get_rand_matrix(data):
    det = 0
    while det == 0:
        matrix = np.random.normal(size=(data.shape[1], data.shape[1]))
        det = np.linalg.det(matrix)
    return matrix
    
```

```python
random_matrix = get_rand_matrix(features_train_matrix)
random_matrix
```

Проверка на обратимость. Произведение матрицы на её обратную должно дать в результате единичную матрицу, а именно:
- Заполнение нолями
- Заполнение единицами по главной диагонали

```python
np.round(random_matrix @ np.linalg.inv(random_matrix))
```

### Кодирование и декодирование

```python
# Кодирование выполняем произведением матриц
features_train_matrix_encoded = features_train_matrix @ random_matrix
features_test_matrix_encoded = features_test_matrix @ random_matrix
```

```python
# Проверка закодированных данных
features_train_matrix_encoded
```

Данные успешно закодированы, персональные данные под защитой.

```python
# Декодируем
features_train_matrix_decoded = (features_train_matrix_encoded 
                                 @ np.linalg.inv(random_matrix))

# Округление до целых после математических преобразований
features_train_matrix_decoded = np.abs(np.round(features_train_matrix_decoded))

# Возвращаем к представлению исходного датасета
features_train_decoded = pd.DataFrame(
    features_train_matrix_decoded,
    columns=features_train.columns,
    index=features_train.index
)
features_train_decoded[['gender', 'family_members']] = features_train_decoded[['gender', 'family_members']].astype('category')
```
